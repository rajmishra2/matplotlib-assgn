Q2:what is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

A scatter plot (scatter graph) uses dots to represent values for two different numeric variables on x and y axis




```python

import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('this is title')
plt.scatter(x,y)
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

1-subplots method provides a way to plot multiple plots on a single figure.




```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])

plt.subplot(2, 3, 1)
plt.plot(x, y)

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 3, 2)
plt.plot(x, y)

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 3, 3)
plt.plot(x, y)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 4, 5)
plt.plot(x, y)
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

Bar plots are a type of data visualization used to represent data in the form of rectangular bars.


```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
```

```python
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot the data in each subplot
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Subplot 1')

axs[0, 1].plot(x, y2)
axs[0, 1].set_title('Subplot 2')
axs[1, 0].plot(x, y3)
axs[1, 0].set_title('Subplot 3')

axs[1, 1].plot(x, y4)
axs[1, 1].set_title('Subplot 4')

plt.tight_layout()  #tight_layout automatically adjusts subplot params so that the subplot(s) fits in to the figure area

plt.show()
```
